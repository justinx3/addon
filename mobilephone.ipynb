```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('mobile_price_classification.csv')
```

```python
x=data[['battery_power', 'bluetooth', 'clock_speed', 'dual_sim', 'front_cam',
       '4G', 'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'primary_camera',
       'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']]
y=data['price_range']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(32,activation='tanh'))
model.add(Dense(32,activation='tanh'))
model.add(Dense(32,activation=''))
model.add(Dense(32,activation='softmax'))
```

```python
model.compile(optimizer='nadam',loss='binnary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs =20)
```

```python
y_pred =np.argmax(model.predict(x_test),axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
